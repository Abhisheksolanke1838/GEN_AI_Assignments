```python
import random
from collections import defaultdict
```

```python
text = """
Artificial intelligence is amazing.
Artificial intelligence is the future.
AI helps humans solve problems.
AI is used in many applications.
AI is transforming the world.
"""

words = text.lower().split()

print("Total words:", len(words))
print("First 15 words:", words[:15])
```

```python
markov_chain = defaultdict(list)

for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]
    markov_chain[current_word].append(next_word)

print("✅ Markov Chain built successfully!")
print("Example:", list(markov_chain.items())[:5])
```

```python
def generate_text(chain, start_word, length=30):
    word = start_word
    result = [word]

    for _ in range(length - 1):
        if word not in chain:
            break
        word = random.choice(chain[word])
        result.append(word)

    return " ".join(result)
```

```python
print("Generated Text:\n")
print(generate_text(markov_chain, start_word="artificial", length=30))
```

```python
random_start = random.choice(words)

print("Random start word:", random_start)
print("\nGenerated Text:\n")
print(generate_text(markov_chain, start_word=random_start, length=30))
```
