```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
```

```python
# Generate sample dataset with 3 clusters
X, y = make_blobs(n_samples=600, centers=3, cluster_std=1.2, random_state=42)

print("Dataset shape:", X.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=15)
plt.title("Original Training Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

print("✅ GMM Model Trained Successfully!")
```

```python
print("Means of each Gaussian:\n", gmm.means_)
print("\nWeights (mixing probabilities):\n", gmm.weights_)
print("\nCovariances:\n", gmm.covariances_)
```

```python
labels = gmm.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, s=15)
plt.title("Cluster Prediction by GMM")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
generated_data, generated_labels = gmm.sample(300)

print("Generated Data Shape:", generated_data.shape)
```

```python
plt.scatter(generated_data[:, 0], generated_data[:, 1], s=15)
plt.title("Generated Synthetic Data using GMM")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], s=15, label="Real Data")
plt.scatter(generated_data[:, 0], generated_data[:, 1], s=15, label="Generated Data")
plt.title("Real vs Generated Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
