```python
import os
from dotenv import load_dotenv
load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
print("API Loaded:", OPENAI_API_KEY is not None)
```

```python
from openai import OpenAI

client = OpenAI(api_key=OPENAI_API_KEY)

def ask_llm(prompt, model="gpt-4o-mini", temperature=0.4):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role":"user","content":prompt}
        ],
        temperature=temperature
    )
    return response.choices[0].message.content
```

```python
problem = """
A student wants to study 4 subjects in 7 days.
Subjects: ML, DBMS, OS, CN
Each day they can study max 6 hours.
Make an effective study plan with revision.
"""
print(problem)
```

```python
zero_shot_prompt = f"""
You are a helpful AI assistant.
Task: {problem}
Generate a proper schedule in table form.
"""

zero_shot_output = ask_llm(zero_shot_prompt)
print(zero_shot_output)
```

```python
few_shot_prompt = f"""
You are a helpful AI assistant.

Example:
Task: A student wants to study 2 subjects in 3 days, max 5 hours/day.
Output:
Day 1: Subject1 - 3h, Subject2 - 2h
Day 2: Subject1 - 2h, Subject2 - 3h
Day 3: Revision - 5h

Now solve this:
Task: {problem}
Generate a schedule in similar format.
"""

few_shot_output = ask_llm(few_shot_prompt)
print(few_shot_output)
```

```python
interview_prompt = f"""
You are an AI assistant using Interview Prompting.

Step 1: Ask 5 important questions to the student before making the study plan.
Step 2: Assume reasonable answers yourself.
Step 3: Based on assumed answers, generate final study plan.

Task: {problem}
"""

interview_output = ask_llm(interview_prompt)
print(interview_output)
```

```python
cot_prompt = f"""
You are a smart assistant.

Task: {problem}

Think step by step.
1) Break down time available
2) Distribute hours logically
3) Include revision + mock test
Finally output clean schedule
"""

cot_output = ask_llm(cot_prompt)
print(cot_output)
```

```python
tot_prompt = f"""
You are a smart assistant using Tree of Thought Prompting.

Task: {problem}

Step 1: Create 3 different possible study plans (Plan A, Plan B, Plan C).
Step 2: Give pros/cons of each plan.
Step 3: Select the best plan and justify.
Step 4: Output the final chosen plan clearly.
"""

tot_output = ask_llm(tot_prompt)
print(tot_output)
```

```python
print("✅ ZERO SHOT OUTPUT:\n")
print(zero_shot_output)

print("\n\n✅ FEW SHOT OUTPUT:\n")
print(few_shot_output)
```

```python
results = {
    "Zero Shot": zero_shot_output,
    "Few Shot": few_shot_output,
    "Interview Approach": interview_output,
    "Chain of Thought (CoT)": cot_output,
    "Tree of Thought (ToT)": tot_output
}

for k, v in results.items():
    print("="*80)
    print(k)
    print("="*80)
    print(v[:800])  # showing first 800 chars
    print("\n")
```
